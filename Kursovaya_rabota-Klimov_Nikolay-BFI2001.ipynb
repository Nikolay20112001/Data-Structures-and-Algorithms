Но я не хочу использовать слишком много чисел, поэтому я выберу три целых положительных числа (необязательно различных), каждое из которых не превышает n. Помогите мне найти максимально возможное наименьшее общее кратное этих трех целых чисел.

Входные данные\
В первой строке записано целое число n (1 ≤ n ≤ 106) — переменная n из условия.

Выходные данные\
Выведите единственное целое число — максимально возможное наименьшее общее кратное трех необязательно различных целых чисел, которые не превышают n.

Примеры\
входные данные\

выходные данные\

Примечание\
Наименьшее общее кратное нескольких положительных целых чисел — это наименьшее положительное целое число, кратное им всем.

Результат может получиться достаточно большим. Возможно, 32-битного целого числа не будет достаточно для его хранения. Поэтому рекомендуется использовать 64-битные целые числа.

В последнем примере мы можем выбрать числа 7, 6, 5, их НОК равен 7·6·5 = 210. Это — максимальный НОК, который мы можем получить.

```python
def correct_n(n):
    if n.isdigit() and int(n) <= 106 and int(n) >=1:
        return int(n)
    else:
        return -1

def maxNOK(n):    
    if n < 3:
        return str(n)
    elif n % 2 == 1:
        return str(n * (n - 1) * (n - 2))
    elif n % 3 == 1:
        return str(n * (n - 1) * (n - 3))
    else:
        return str((n - 1) * (n - 2) * (n - 3))
    
def main():
    n = input("Введите число n: ")
    n = correct_n(n)
    if n != -1:
        print("Максимальное НОК равен: " + maxNOK(n))
    else:
        print("Неверное значение n!")

main()
```

gcd(𝑠) это максимальное натуральное число 𝑥, такое что все числа из 𝑠 делятся на 𝑥.\
lcm(𝑠) это минимальное натуральное число 𝑥, которое делится на все числа из 𝑠.\
Например, gcd({8,12})=4,gcd({12,18,6})=6 и lcm({4,6})=12. Обратите внимание, что для любого натурального числа 𝑥, gcd({𝑥})=lcm({𝑥})=𝑥.

Входные данные\
В первой строке записано одно целое число 𝑛 (2≤𝑛≤100000).

Во второй строке записаны 𝑛 целых чисел, 𝑎1,𝑎2,…,𝑎𝑛 (1≤𝑎𝑖≤200000).

Во втором примере 𝑡={120,40,80,120,240,80}. Нетрудно видеть, что gcd(𝑡)=40.

```python
def correct_n(n):
    if n.isdigit() and int(n) <= 100000 and int(n) >=2:
        return int(n)
    else:
        return -1

def massiv_s_input(n):
    massiv_s_input = []
    massiv_s_input = input("Введите " + str(n) + " целых чисел: ")
    massiv_s_input = massiv_s_input.split()
    return massiv_s_input

def correct_massiv_s(massiv_s,n):
    counter_correct_input = 0
    if len(massiv_s) == n:
        for i in range(n):
            if massiv_s[i].isdigit() and int(massiv_s[i]) >= 1 and int(massiv_s[i]) <= 200000:
                counter_correct_input += 1
        if (counter_correct_input == n):
            for i in range(n):
                massiv_s[i] = int(massiv_s[i])
            return  massiv_s
        else:
            return -1
    else:
        return -1

def search_NOKs(s):
    massiv_of_NOKs = []
    max_number_couple = 0
    min_number_couple = 0
    step_search_NOK = 0
    flag = 0
    for i in range(len(s)):
        for j in range(i+1,len(s)):
            step_search_NOK = max(s[i],s[j])
            max_number_couple = step_search_NOK
            min_number_couple = min(s[i],s[j])
            while flag == 0:
                if max_number_couple % min_number_couple == 0 and massiv_of_NOKs.count(max_number_couple) == 0:
                    massiv_of_NOKs.append(max_number_couple)
                    flag = 1
                elif max_number_couple % min_number_couple == 0 and massiv_of_NOKs.count(max_number_couple) != 0:
                    flag = 1
                else:
                    max_number_couple += step_search_NOK
            step_search_NOK = 0
            min_number_couple = 0
            max_number_couple = 0
            flag = 0
    return massiv_of_NOKs

def search_HOD_of_NOKs(s):
    max_del = 0
    current_del = 0
    counter_every_del = 0
    
    massiv_of_NOKs = search_NOKs(s)
    
    current_del = min(massiv_of_NOKs)
    while current_del != 0:
        for i in massiv_of_NOKs:
            if i % current_del == 0:
                counter_every_del += 1
        if counter_every_del == len(massiv_of_NOKs):
            max_del = current_del
            break
        else:
            counter_every_del = 0
            current_del -= 1
    return max_del

def main():
    n = input("Введите длину последовательности: ")
    n = correct_n(n)
    s = []
    if n != -1:
        s = massiv_s_input(n)
        if correct_massiv_s(s,n) != -1:
            print("НОД НОКов всех пар равен: " + str(search_HOD_of_NOKs(s)))
        else:
            print("Некорректный ввод массива чисел!")
    else:
        print("Неверное значение длины последовательности!")
        
main()
```

Единственное, что Василий может делать, это разворачивать строки (первая буква становится последней, вторая предпоследней и так далее).

Чтобы развернуть i-ю строку Василию, надо потратить ci единиц энергии. Василия интересует минимальное количество энергии, которое необходимо потратить, чтобы строки шли в лексикографическом порядке.

Строка A лексикографически меньше строки B, если она короче B (|A| < |B|) и является её префиксом, либо ни одна из них не является префиксом другой, и в первой позиции, где они различаются, в строке A стоит символ с меньшим номером.

В данной задаче две одинаковые строки на соседних позициях не нарушают порядок лексикографической сортировки.

Входные данные\
В первой строке входных данных записано единственное целое число n (2 ≤ n ≤ 100 000) — количество строк.

Во второй строке записаны n целых чисел ci (0 ≤ ci ≤ 109), i-е из которых равняется количеству энергии, необходимому Василию для разворота i-й строки.

Далее следуют n строк, состоящих из строчных букв английского алфавита. Суммарная длина всех строк не превышает 100 000.

Выходные данные\
Если, разворачивая какие-либо из данных строк, невозможно добиться, чтобы они следовали в лексикографическом порядке, то выведите  - 1. В противном случае выведите минимальное количество энергии, которое придётся потратить Василию.

Примеры\
входные данные\
2\
1 2\
ba\
ac\
выходные данные\
1\
входные данные\
3\
1 3 1\
aa\
ba\
ac\
выходные данные\
1\
входные данные\
2\
5 5\
bbb\

выходные данные\
-1\
входные данные\

3 3\
aaa\
aa\
выходные данные\
-1\
Примечание\
Во втором примере можно развернуть строку 2 или строку 3. На разворот строки 3 тратится меньше энергии, поэтому правильным ответом будет развернуть её. В третьем примере обе строки не изменяются после разворота и расположены в неправильном порядке, поэтому ответом является  - 1. В четвёртом примере обе строки состоят из букв «a», но в отсортированном порядке строка «aa» должна располагаться раньше строки «aaa», поэтому ответ  - 1.

```python
def correct_n(n):
    if n.isdigit() and int(n) <= 100000 and int(n) >=2:
        return int(n)
    else:
        return -1

def input_ci():
    return input("Введите количество энергии для разворота i-ой строки: ").split()

def input_strings(n):
    strings = []
    for i in range(n):
        strings.append(input("Введите строку: "))
    return strings

def correct_ci(ci,n):
    if len(ci) != n:
        return -1
    counter_correct_input = 0
    for i in range(n):
        if ci[i].isdigit() and int(ci[i]) >= 0 and int(ci[i]) <= 109:
            counter_correct_input += 1
    if (counter_correct_input == n):
        for i in range(n):
            ci[i] = int(ci[i])
        return  ci
    else:
        return -1

def correct_strings(strings,n):
    counter_correct_lc = 0
    counter_correct_input = 0
    for i in range(n):
        for j in range(len(strings[i])):
            if strings[i][j].isalpha() and strings[i][j].islower():
                counter_correct_lc += 1
        if len(strings[i]) == counter_correct_lc:
            counter_correct_input += 1
            counter_correct_lc = 0
        else:
            return -1
    if (counter_correct_input == n):
        return  strings
    else:
        return -1
    
def energy_strings(n,ci,strings):
    start_and_end = []
    result = 0
    current_string = 0
    flags = []
    for i in range(len(ci)):
        start_and_end.append(list(strings[i][::]))
        flags.append(0)

    while current_string < len(strings) - 1:
        for i in range(min(len(start_and_end[current_string]),len(start_and_end[current_string+1])) - 1):
            if ord(start_and_end[current_string][i]) > ord(start_and_end[current_string+1][i]) and ci[current_string] < ci[current_string + 1] and flags[current_string] != 1:
                start_and_end[current_string].reverse()
                flags[current_string] = 1
                result += ci[current_string]

            if ord(start_and_end[current_string][i]) > ord(start_and_end[current_string+1][i]) and ci[current_string] > ci[current_string + 1]  and flags[current_string] != 1:
                start_and_end[current_string+1].reverse()
                flags[current_string + 1] = 1
                result += ci[current_string + 1]
            elif not(ord(start_and_end[current_string][i]) <= ord(start_and_end[current_string+1][i])):
                result = -1
                break
        if "".join(map(str,start_and_end[current_string])).find("".join(map(str,start_and_end[current_string + 1]))) != -1 and len(start_and_end[current_string]) > len(start_and_end[current_string+1]):
            result = -1
            break
        current_string += 1
    return result

def main():
    n = input("Введите количество строк: ")
    n = correct_n(n)
    if n != -1:
        ci = input_ci()
        strings = input_strings(n)
        if correct_ci(ci,n) != -1 and correct_strings(strings,n) != -1:
            print("Минимальное количество энергии: " + str(energy_strings(n,ci,strings)))
        else:
            print("Неверный ввод значений энергии или строк!")
    else:
        print("Неверное значение количества строк!")
        
main()
        
```

### Задача 10
У Феникса есть строка 𝑠, состоящая из строчных букв латинского алфавита. Он хочет распределить все буквы своей строки по 𝑘 непустым строкам 𝑎1,𝑎2,…,𝑎𝑘 так, что каждая буква из 𝑠 попадет ровно в одну из строк 𝑎𝑖. Строки 𝑎𝑖 не обязаны быть подстроками 𝑠. Феникс может распределить буквы 𝑠 и переупорядочить их внутри каждой строки 𝑎𝑖 так как захочет.

Например, если 𝑠= baba и 𝑘=2, Феникс может распределить буквы своей строки множеством способов, в том числе:

ba и ba

Однако получить такие варианты он не может:

baa и ba

baba и пустая строка (𝑎𝑖 должны быть непустыми)
Феникс хочет разделить свою строку 𝑠 на 𝑘 строк 𝑎1,𝑎2,…,𝑎𝑘 так, чтобы минимизировать лексикографически максимальную строку среди них, т. е. минимизировать 𝑚𝑎𝑥(𝑎1,𝑎2,…,𝑎𝑘). Помогите ему найти оптимальное распределение и выведите минимально возможное значение 𝑚𝑎𝑥(𝑎1,𝑎2,…,𝑎𝑘).

Строка 𝑥 лексикографически меньше, чем строка 𝑦, если либо 𝑥 является префиксом 𝑦 (и 𝑥≠𝑦), либо существует такой индекс 𝑖 (1≤𝑖≤𝑚𝑖𝑛(|𝑥|,|𝑦|)), что 𝑥𝑖 < 𝑦𝑖 и для всех 𝑗 (1≤𝑗<𝑖) 𝑥𝑗=𝑦𝑗. Здесь |𝑥| обозначает длину строки 𝑥.

Входные данные\
Входные данные состоят из нескольких наборов. В первой строке задано целое число 𝑡 (1≤𝑡≤1000) — количество наборов входных данных. Каждый набор состоит из двух строк.

В первой строке каждого набора задано два целых числа 𝑛 и 𝑘 (1≤𝑘≤𝑛≤105)  — длина строки 𝑠 и количество не пустых строк, в которые Феникс хочет распределить буквы 𝑠, соответственно.

Во второй строке каждого набора задана строка 𝑠 длины 𝑛, состоящая из строчных латинских букв.

Гарантируется, что сумма 𝑛 по всем наборам входных данных ≤105.

Выходные данные\
Выведите 𝑡 ответов — по одному на набор входных данных; 𝑖-й ответ — минимально возможный 𝑚𝑎𝑥(𝑎1,𝑎2,…,𝑎𝑘) в 𝑖-м наборе.

Пример\
входные данные\

Примечание\
В первом наборе входных данных, одно из оптимальных решений — разбить baba на ab и ab.

Во втором наборе входных данных, одно из оптимальных решений — разбить baacb на abbc и a.

В третьем наборе, одно из оптимальных решений — разбить baacb на ac, ab и b.

В шестом наборе, одно из оптимальных решений — разбить phoenix на ehinopx.

```python
def correct_t(t):
    if t.isdigit() and int(t) <= 1000 and int(t) >=1:
        return int(t)
    else:
        return -1

def correct_n_and_k(n_and_k):
    counter_correct_input = 0
    if len(n_and_k) > 2:
        return -1
    if n_and_k[0].isdigit() and int(n_and_k[0]) <= 105 and int(n_and_k[0]) >= 1:
        counter_correct_input += 1
    else:
        return -1
    if n_and_k[0].isdigit() and int(n_and_k[1]) <= 105 and int(n_and_k[1]) >= 1 and int(n_and_k[0]) >= int(n_and_k[1]):
        counter_correct_input += 1
    else:
        return -1
    
    if (counter_correct_input == 2):
        for i in range(2):
            n_and_k[i] = int(n_and_k[i])
        return  n_and_k
    else:
        return -1
    
def correct_a(a):
    counter = 0
    if a.count(' ') != 0:
        return -1
    for i in range(len(a)):
        if a[i].islower() and a[i].isalpha():
            counter += 1
    if counter == len(a):
        return a
    else:
        return -1

def minmax_combination(t,n,k,a):
    a = list(a)
    a.sort()
    result1 = [''] * k
    result2 = [''] * k
    for i in range(n):
        result1[i % k] += a[i]
    for i in range(k):
        result2[i] += a[i]
    for i in range(k, n):
        result2[0] += a[i]
    return min(max(result1), max(result2))

def main():
    t = input("Введите количество наборов входных данных: ")
    n_and_k = input("Введите n и k: ").split()
    a = input("Введите набор символов: ")
    t = correct_t(t)
    if t != -1:
        if correct_n_and_k(n_and_k) != -1:
            if correct_a(a) != -1:
                for i in range(t):
                    print("Результат: " + str(minmax_combination(t,n_and_k[0],n_and_k[1],a))
            else:
                print("Последовательность a неверно введена!")
        else:
            print("Некорректный ввод длины строки или количества не пустых строк")
    else:
        print("Неверное значение количества тестовых наборов!")
                          
main()
```
