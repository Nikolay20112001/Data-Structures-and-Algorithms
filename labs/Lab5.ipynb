```python
#Кривая Коха
import turtle

def koch_curve(t,length):
    if length>6:
        length_n = length//3
        koch_curve(t,length_n)
        t.left(60)
        koch_curve(t,length_n)
        t.right(120)
        koch_curve(t,length_n)
        t.left(60)
        koch_curve(t,length_n)        
    else:
        t.forward(length)
        t.left(60)
        t.forward(length)
        t.right(120)
        t.forward(length)
        t.left(60)
        t.fd(length)        

width = 1200
heidht = 800
screen = turtle.Screen()
screen.setup(width, heidht,0,0)

t = turtle.Turtle()
t.ht()
t.speed(100)
koch_curve(t,150)
turtle.done()
```

```python
#Кривая Гильберта
import turtle           

def A(size, n):
    if n==0:          
        return
    t.left(90)        
    B(size, n-1)     
    t.fd(size)        
    t.right(90)      
    A(size, n-1)     
    t.fd(size)        
    A(size, n-1)      
    t.right(90)      
    t.fd(size)        
    B(size, n-1)     
    t.left(90)        

def B(size, n):
    if n==0:          
        return
    t.right(90)       
    A(size, n-1)       
    t.fd(size)        
    t.left(90)        
    B(size, n-1)    
    t.fd(size)        
    B(size, n-1)      
    t.left(90)       
    t.fd(size)        
    A(size, n-1)           
    t.right(90)      
    
t = turtle.Turtle()
t.shape("turtle")
t.ht()
t.width(3)
t.speed(100)

A(40, 3)     

t.hideturtle()
turtle.done()
```

```python
#Кривая Серпинского
import turtle

def serp_out(size, depth):
    serp(size, depth, "end")
    t.right(90 + 45)
    t.forward(size/pow(2, depth))

def serp (size, depth, position):
    if depth == 0:
        t.left(45)
        t.fd(size)
        t.right(90)
        t.fd(size)
        t.right(90)
        t.fd(size)
        t.left(45)
    else:
        if position == 'diag up': 
            if depth != 1:
                serp(size/2, depth-1, "diag up")
            t.left(45)
            t.fd(size/pow(2, depth))
            t.left(45)
            t.fd(size/pow(2, depth-1))
            serp(size, depth, "end")
            t.fd(size/pow(2, depth-1))
            if depth != 1:
                serp(size/2, depth-1, "diag up")
        elif position == 'diag down':
            if depth != 1:
                serp(size/2, depth-1, "diag down")
            t.forward(size/pow(2, depth-1))
            serp(size, depth, "end")
            t.forward(size/pow(2, depth-1))
            t.left(45)
            t.forward(size/pow(2, depth))
            t.left(45)
            if depth != 1:
                serp(size/2, depth-1, "diag down")
        elif position == 'center':
            t.left(45)
            t.forward(size/pow(2, depth))
            t.left(45)
            t.forward(size/pow(2, depth-1))
            serp(size, depth, "end")
            t.forward(size/pow(2, depth-1))
            serp(size, depth, "end")
            t.forward(size/pow(2, depth-1))
            serp(size, depth, "end")
            t.forward(size/pow(2, depth-1))
            t.left(45)
            t.forward(size/pow(2, depth))
            t.left(45)
            
        else:
            if depth > 2:
                serp(size/4, depth-2, "diag up")
            serp(size/2, depth-1, "center")
            if depth > 2:
                serp(size/4, depth-2, "diag down")
                
size = 100
depth = 4
t = turtle.Turtle()
t.clear()
t.setheading(0)
t.speed(10)

serp_out(size, depth)
turtle.mainloop()
```

```python
#Салфетка Серпинского
import turtle
def A(size, depth):
    if depth==0:          
        return
    B(size, depth-1)      
    t.fd(size)       
    t.left(60)       
    A(size, depth-1)      
    t.fd(size)       
    t.left(60)       
    B(size, depth-1)      

def B(size, depth):
    if depth==0:          
        return
    A(size, depth-1)     
    t.fd(size)       
    t.right(60)       
    B(size, depth-1)     
    t.fd(size)       
    t.right(60)      
    A(size, depth-1)     
    
t = turtle.Turtle()
t.shape("turtle")
t.ht()
t.width(3)
t.speed(100)

A(10, 5)    

t.hideturtle()
turtle.done()
```

```python
#Ковер Серпинского
import turtle

def s(l,depth):
    if depth == 0:
        t.color('black')
        t.begin_fill()
        for i in range (4):
            t.fd(l)
            t.left(90)
        t.end_fill()

    else:
        for i in range(4):
            s(l/3,depth-1)    
            t.fd(l/3)
            s(l/3,depth-1)     
            t.fd(l/3)
            t.fd(l/3)
            t.left(90)
        turtle.update()
t = turtle.Turtle()
t.ht()
t.speed(100)

s(400, 3)
turtle.done()
```

```python
#Дерево Пифагора
import turtle
import math
def tree_of_Pifagor(size, depth):
    if depth == 0:
        t.left(90)
        for i in range(0,3):
            t.fd(size*math.sqrt(2)/2)
            t.right(90)
        t.fd(size*math.sqrt(2)/2)
    else:
        t.left(90)
        t.fd(size)
        t.right(45)
        if depth == 1:
            tree_of_Pifagor(size, depth-1)
        else:
            tree_of_Pifagor(size*math.sqrt(2)/2, depth-1)
        t.right(90)
        t.right(135)
        t.fd(size)
        t.left(45)
        if depth == 1:            
            tree_of_Pifagor(size, depth-1)
            t.right(90)
            t.right(225)
        else:
            t.left(90)
            t.fd(size*math.sqrt(2)/2)
            t.right(180)
            tree_of_Pifagor(size*math.sqrt(2)/2, depth-1)
            t.right(180)
            t.forward(size*math.sqrt(2)/2)
            t.right(45)
        t.fd(size)
        t.right(90)
        t.fd(size)
def sierpinski_half(size, level):
    if level == 0: 
        t.fd(size)
        return
    sierpinski_half(size, level - 1)
    t.left(45)
    t.fd(size*math.sqrt(2)) 
    t.left(45)
    sierpinski_half(size, level - 1)
    t.right(90)
    t.fd(size)
    t.right(90)
    sierpinski_half(size, level - 1)
    t.left(45)
    t.fd(size*math.sqrt(2)) 
    t.left(45)
    sierpinski_half(size, level - 1)

def line_sierpinski(size, level):
    sierpinski_half(size, level)
    t.right(90)
    t.fd(size)
    t.right(90)
    sierpinski_half(size, level)
    t.right(90)
    t.fd(size)
    t.right(90)
t = turtle.Turtle()
t.shape("turtle")
t.ht()
t.width(3)
t.speed(100)
tree_of_Pifagor(100,3)
t.hideturtle()
turtle.done()
```
